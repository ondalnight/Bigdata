```python
import numpy as np

a = np.arange(1, 13).reshape(3, 4)
c = np.arange(1, 6, 2)

a, c
```

```python
c[:, None]
```

```python
c[:, np.newaxis]
```

```python
cc = np.expand_dims(c, axis = 1)
```

```python
np.hstack((a, cc))
```

```python
np.column_stack((a, c))
```

```python
np.column_stack((a, cc))
```

```python
a = np.arange(1, 4)
b = np.arange(4, 7)
a, b
```

```python
np.c_[a, b], np.c_[(a, b)]
```

```python
np.c_[a, b], np.column_stack((a, b))
```

```python
x = np.arange(1, 4).reshape(1, 3)
y = np.arange(4, 7).reshape(1, 3)
x, y
```

```python
np.c_[x, 10, 20, y]
```

```python
np.column_stack((x, 10, 20, y))
```
