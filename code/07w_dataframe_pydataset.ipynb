```python
from IPython.display import display_html
def display_side_by_side(*args):
    """여러 데이터프레임 비교가 쉽게 옆쪽으로 표시한다"""
    html_str=''
    for df in args:
        html_str += df.to_html() + '&nbsp;'*4
    display_html(html_str.replace('table','table style="display:inline"'), raw=True)
```

```python
display_side_by_side(df.fillna(10), df)
```

```python
import os
import pandas as pd

# 현재 디렉토리 파일 목록 확인
files = os.listdir('.')
target_files = [f for f in files if '2022' in f and f.endswith('.csv')]

if target_files:
    filename = target_files[0]
    # header=1로 실제 컬럼명을 가져오고, thousands=','를 사용하여 숫자의 콤마를 제거하며 읽어옵니다.
    traffic = pd.read_csv(filename, encoding='cp949', header=1, thousands=',')

    # 컬럼명 공백 제거
    traffic.columns = [col.strip() for col in traffic.columns]

    display(traffic.head())
    print("\n현재 컬럼 정보:")
    print(traffic.info())
else:
    print("파일을 찾을 수 없습니다.")
```
